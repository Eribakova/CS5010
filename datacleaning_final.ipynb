```python
temp = df2[["start_station_id","new_start_lat","new_start_lng"]]
temp = temp.groupby("start_station_id").mean()
temp= temp[temp['new_start_lat'].notnull()]
temp
```

```python
startlatlst=temp['new_start_lat'].tolist()
startlnglst=temp['new_start_lng'].tolist()
newlst=[]
x=list((zip(startlatlst,startlnglst)))
for i in x:
  x0=round(i[0],6)
  x1=round(i[1],6)
  newpair=str(x0) + ","+str(x1)
  newlst.append(newpair)
newlst = list( dict.fromkeys(newlst))
print(len(newlst))
```

```python
print(newlst[0:10])
```

```python
statels=[]
for i in newlst:
  try:
    location=(geolocator.reverse(i))
    address = location.raw['address']
    state = address.get('state', '')
    statels.append(state)
  except:
    print('error in',i)
    break
temp['state']=statels
```

```python
# County Variable
#Start a log for error reporting

fileout=open("Exception_Log.txt",'a')

#Create County variable
countyls=[]
for i in newlst:
  try:
    location=(geolocator.reverse(i))
    address = location.raw['address']
    county = address.get('county', '')
    countyls.append(county)
  except:
    print('error in',i)
    fileout.write(i)
    break

temp['county']=countyls
```

```python
#Create City variable
cityls=[]
for i in newlst:
  try:
    location=(geolocator.reverse(i))
    address = location.raw['address']
    city = address.get('city', '')
    cityls.append(city)
  except:
    print('error in',i)
    fileout.write(i)

    break
temp['city']=cityls

fileout.close()
```

```python
#This section combines county and city to create a single 'locale' variable.
#https://www.codegrepper.com/code-examples/python/pandas+replace+column+values+with+another+column
temp['locale']=temp['county']
temp['locale'] = np.where(temp['locale'] == "", temp['city'], temp['locale'])
```

```python
covid1 = covid[['Report Date','Total Cases','Hospitalizations','Deaths']]
covid
```

```python
covid_counts = covid1.groupby(['Report Date']).sum().sort_values('Report Date',ascending=False)
covid_counts=covid_counts.reset_index()
covid_counts
```

```python
covid_counts["Report Date"]= pd.to_datetime(covid_counts['Report Date'])
covid_counts
```

```python
# Converting 'date' to merge with Covid 'Report Date'
df3["date"]= pd.to_datetime(df3["date"])
df3
```

```python
df4 = pd.merge(left=df3, right=covid_counts, how='left', 
               left_on='date', right_on='Report Date')
```

```python
df4['Total Cases'] = df4['Total Cases'].replace(np.nan, 0)
df4['Hospitalizations'] = df4['Hospitalizations'].replace(np.nan, 0)
df4['Deaths'] = df4['Deaths'].replace(np.nan, 0)
df4
```

```python
df4.sort_values(by='date')
```

```python
# Calculating duration to fill out missing values and converting into seconds
df4["started_at"]=pd.to_datetime(df4["started_at"])
df4["ended_at"]=pd.to_datetime(df4["ended_at"])
df4["new_duration"] = df4["ended_at"] - df4["started_at"]
df4['new_duration(min)'] = df4['new_duration'].dt.total_seconds()
df4
```
